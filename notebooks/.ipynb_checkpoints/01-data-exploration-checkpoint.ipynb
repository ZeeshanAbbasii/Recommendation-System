### Understanding the `helpful` Column in Amazon Reviews Dataset

In Amazon reviews datasets, the `helpful` column typically looks like this:

It means:
- **2 people found the review helpful.**
- **3 people voted** (clicked “Yes” or “No”).

So this review has a **helpfulness ratio** of:

2/3 =0.667 which is `67 percent approx`

```python
import pandas as pd
import json

data = []
with open('/Users/zeeshanabbasi/Desktop/RecSys-Final-Project/data/raw/Electronics_5.json', 'r') as f:
    for line in f:
        data.append(json.loads(line))

df = pd.DataFrame(data)
```

```python
print(type(df))
```

```python
df['reviewTime'] = pd.to_datetime(df['reviewTime'])
df.head()
```

```python
df['user_id'] = df['reviewerID'].astype('category').cat.codes
df['item_id'] = df['asin'].astype('category').cat.codes
#converting ASJDLS type code to integers so our algorithm can understand
```

```python
num_users = df['user_id'].nunique()
num_items = df['item_id'].nunique()

print(f"Unique Users: {num_users}")
print(f"Unique Items: {num_items}")
```

```python
df['reviewYear']= df['reviewTime'].dt.year
df['reviewMonth']= df['reviewTime'].dt.month
```

```python
df.to_csv('/Users/zeeshanabbasi/Desktop/RecSys-Final-Project/data/processed/cleaned_reviews.csv', index=False)
```

```python
#visualizing user and product distributions
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
user_review_counts = final_dataset['user_id'].value_counts()
plt.figure(figsize=(10,4))
sns.histplot(user_review_counts, bins=50, kde= False)
plt.title('Distribution of Number of Reviews per user:')
plt.xlabel('Number of Reviews')
plt.ylabel('Number of Users')
plt.xlim(0,100)
plt.show()
```

```python
item_review_counts = final_dataset['item_id'].value_counts()
plt.figure(figsize=(10,4))
sns.histplot(item_review_counts, bins=10, kde=False, color='orange')
plt.title('Distribution of Number of Reviews per item')
plt.xlabel('Number of Reviews')
plt.ylabel('Number of Items')
plt.show()
```

```python
top_products = final_dataset['item_id'].value_counts().head(10) #top 10 products
plt.figure(figsize=(10,5))
sns.barplot(x=top_products.values,y=top_products.index)
plt.title('Top 10 most reviewed products')
plt.xlabel('Number of Reviews')
plt.ylabel('Product ID')
plt.show()
```

```python
#most active users
top_users = final_dataset['user_id'].value_counts().head(10)
plt.figure(figsize=(15,5))
sns.barplot(x=top_users.values,y=top_users.index)
plt.title('Top 10 most active users')
plt.xlabel('Number of Reviews')
plt.ylabel('User ID')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=final_dataset, x='overall')
plt.xlabel('Distribution of Ratings')
plt.ylabel('Ratings')
plt.show()
```

```python
num_users = final_dataset['user_id'].nunique()
num_items = final_dataset['item_id'].nunique()
num_interactions = len(final_dataset)

sparsity = 1 - (num_interactions / (num_users * num_items))
print(f"Sparsity of the dataset: {sparsity:.4f}")
```

```python
# we did sparsity check to see how much users have interacted with how many items, so sparsity close to 1 means that most users have interacted with many items
```

```python
#now seeing time based trends
plt.figure(figsize=(10,5))
sns.countplot(data=final_dataset, x= 'reviewYear')
plt.title('Number of Reviews per year')
plt.xlabel('Year')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
monthly_reviews = final_dataset.groupby(['reviewYear', 'reviewMonth']).size().reset_index(name='count')
monthly_reviews['date'] = pd.to_datetime(monthly_reviews['reviewYear'].astype(str) + '-' + monthly_reviews['reviewMonth'].astype(str))

plt.figure(figsize=(12, 5))
sns.lineplot(data=monthly_reviews, x='date', y='count')
plt.title('Monthly Reviews Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
avg_rating_year= final_dataset.groupby('reviewYear')['overall'].mean().reset_index()
plt.figure(figsize=(10,5))
sns.lineplot(data=avg_rating_year, x='reviewYear',y='overall')
plt.title('Average Rating per year')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.show()
```

```python
heatmap_data = final_dataset.groupby(['reviewYear', 'reviewMonth']).size().unstack(fill_value=0)
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='d')
plt.title('Heatmap of Reviews by Year and Month')
plt.xlabel('Month')
plt.ylabel('Year')
plt.tight_layout()
plt.show()
```

```python
# we have covered most of the EDA part but it will be useful if we get more deep into:
#	•	Products with both high number of reviews and high average rating.
#	•	Products with many reviews but low ratings (maybe “bad” products).
#	•	Is there a correlation between review length and rating?
```

```python
product_stats = final_dataset.groupby('item_id').agg(
    review_count=('overall', 'count'),
    avg_rating=('overall', 'mean')
).reset_index()
top_good_products = product_stats[(product_stats['review_count'] > 100) & (product_stats['avg_rating'] >= 4.5)]
top_good_products.sort_values(by='avg_rating', ascending=False).head(10)
```

```python
bad_popular_products = product_stats[(product_stats['review_count'] > 100) & (product_stats['avg_rating'] <= 3)]
bad_popular_products.sort_values(by='avg_rating').head(10)
```

```python
final_dataset['review_length'] = final_dataset['reviewText'].astype(str).apply(len)
correlation = final_dataset[['review_length', 'overall']].corr()
print("Correlation between review length and rating:")
print(correlation)
```

- Very weak **negative** correlation.
- As review length increases, the rating slightly tends to decrease.
- The relationship is **negligible** and likely not meaningful.
